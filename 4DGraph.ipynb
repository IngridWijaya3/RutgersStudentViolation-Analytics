```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pylab import *

gridcount=pd.read_csv('LocationAverageWeek.csv',keep_default_na=False)
top20location=pd.read_csv('Top20LocationAverageWeek.csv',keep_default_na=False)

top20location["latitude"]=top20location["latitude"].astype(float)
top20location["longitude"]=top20location["longitude"].astype(float)
gridcount["latitude"]=gridcount["latitude"].astype(float)
gridcount["longitude"]=gridcount["longitude"].astype(float)
gridcount = gridcount.reset_index(drop=True)
top20location = top20location.reset_index(drop=True)
gridcount=pd.merge(gridcount, top20location, on =['latitude','longitude'], how='inner')
#gridcount=gridcount.loc[((gridcount["latitude"]==top20location["latitude"]) & (gridcount["longitude"]==top20location["longitude"]))]
data=gridcount.as_matrix()
#1= latitude
#2= longitude
#3=week_number
x = data[:,2]  # select columns 1 through end
y = data[:,1]  
z = data[:,3]  
#X = np.reshape(data[:,0],(100,1))  

c = data[:,4] 


fig = plt.figure(figsize=(20,15))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('longitude')
ax.set_ylabel('latitude')
ax.set_zlabel('Week')
ax.set_xlim3d(-74.45480,-74.43050)
ax.set_ylim3d(40.47840,40.52040)
ax.set_zlim3d(0,53)
#colmap = cm.ScalarMappable(cmap=plt.cm.Reds)
#colmap.set_array(c)

#yg = ax.scatter(xs, ys, zs, c=cm.hsv(zs/max(zs)), marker='o')

ax.scatter(x, y, z, c=c, cmap=plt.cm.Blues)# cmap=plt.hot())
#cb = fig.colorbar(colmap)
plt.show()
#gridcount
```
