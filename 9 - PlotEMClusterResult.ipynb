```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
from sklearn.cluster import KMeans
from sklearn import datasets

np.random.seed(5)


gridcount=pd.read_csv('GridCount.csv',keep_default_na=False)
data=gridcount.as_matrix()

x = data[:,10]
y = data[:,11]
Label = np.sort(np.unique(data[:,14] ))
Cluster =data[:,15]    # Labels of cluster 0 to 3

fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111)
scatter = ax.scatter(x,y,c=Cluster,label=Label)

ax.set_xlabel('Number of Violation')
ax.set_ylabel('Number of Location')
plt.colorbar(scatter)

plt.legend()
plt.show()
```
